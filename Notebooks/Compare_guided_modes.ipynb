```python
import sys
sys.path.append('../')

import numpy as np

%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from itertools import zip_longest
import time
```

```python
# New version, TM polarization
t = time.time()
from pygme.guided_modes_new import guided_modes
gs=np.linspace(0,50,500)
epses=np.array([1,12,1])
ds=np.array([0.1])
(omegas, coeffs) = guided_modes(gs, epses, ds, step=0.1e-1, tol=1e-6, n_modes=4, mode='TM')
omegasT = [list(filter(None, i)) for i in zip_longest(*omegas)]

for w in omegasT:
    plt.plot(gs[-len(w):],w,'.')
plt.show()
print("Done in %1.4fs" % (time.time()-t))
```
