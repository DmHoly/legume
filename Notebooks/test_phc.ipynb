```python
import sys
sys.path.append('../')

import numpy as np
import pygme

%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
```

```python
# Initialize a lattice
lattice = pygme.Lattice('square')
# Initialize a PhC (by default with eps = 1 in upper and lower cladding, we set upper one to 5)
phc = pygme.PhotCryst(lattice, eps_u = 5)
# Add a layer to the PhC with thickness 1 and background permittivity 10
phc.add_layer(d=1, eps_b=10)
# Add a shape to this layer 
circle = pygme.Circle(r=0.2)
phc.add_shape(circle)
# Add another layer and another shape
phc.add_layer(d=0.5, eps_b=6)
phc.add_shape(pygme.Circle(x_cent=0.3, y_cent=0, r=0.3, eps=12), layer=1)
# Plot an xz cross section of the permittivity
phc.plot_cross('xz')
# Plot an overview picture
phc.plot_overview()
```

```python
# We can also add shapes to claddings
circle = pygme.Circle(x_cent=0.3, y_cent=0, r=0.3, eps=10)
phc.add_shape(circle, cladding='l')
phc.plot_overview(cladding=True)
# Need to improve the plotting...
```
