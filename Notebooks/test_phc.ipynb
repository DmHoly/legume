```python
import numpy as np
from phc import PhotCryst, Lattice

%load_ext autoreload
%autoreload 2
```

```python
# Initialize a lattice
lattice = Lattice('square')
# Initialize a PhC (by default with eps = 1 in upper and lower cladding, we set upper one to 5)
phc = PhotCryst(lattice, eps_u = 5)
# Add a layer to the PhC with thickness 1 and background permittivity 10
phc.add_layer(d=1, eps_b=10)
# Add a shape to this layer 
phc.add_shape('circle', {'x': 0, 'y': 0, 'r': 0.2, 'eps': 1}, layer_ind=0)
# Add another layer and another shape
phc.add_layer(d=0.5, eps_b=6)
phc.add_shape('circle', {'x': 0.3, 'y': 0, 'r': 0.3, 'eps': 12}, layer_ind=1)
# Plot an xz cross section of the permittivity
phc.plot_cross('xz')
# Plot an overview picture
phc.plot_overview()
```

```python
# We can also do polygonal shapes
phc = PhotCryst(lattice)
phc.add_layer(d=1, eps_b=10)

# Square; note that both PhotCryst and Layer classes have add_shape() methods
phc.layers[-1].add_shape('square', {'x_cent': 0.2, 'y_cent': 0.5, 'a': 0.3, 'eps': 2})
phc.plot_overview()

# Arbitrary polygon 
phc.add_layer(d=0.4, eps_b=8)
phc.layers[-1].add_shape('poly', {'x_edges': [-0.1, 0.5, 0.2], 'y_edges': [0, 0, 0.5], 'eps': 4})
phc.plot_overview()
```

```python
# Let's try a hexagonal lattice
lattice = Lattice('hexagonal')
# Initialize a PhC (by default with eps = 1 in upper and lower cladding, we set upper one to 5)
phc = PhotCryst(lattice, eps_u = 5)
# Add a layer to the PhC with thickness 1 and background permittivity 10
phc.add_layer(d=1, eps_b=10)
# Add a circle and a square
phc.add_shape('circle', {'x': 0, 'y': 0, 'r': 0.2, 'eps': 1}, layer_ind=0)
phc.layers[-1].add_shape('square', {'x_cent': 0, 'y_cent': 0.4, 'a': 0.1, 'eps': 2})
# Add another layer and another shape
phc.add_layer(d=0.5, eps_b=6)
phc.add_shape('circle', {'x': 0.3, 'y': 0, 'r': 0.3, 'eps': 12}, layer_ind=1)
# Plot an xz cross section of the permittivity
phc.plot_overview()
```

```python
print(phc.lattice.b1, phc.lattice.b2)
```
